###### Loading dataset

****Loading Atots for timeseries data****

```python
from AutoTS.AutoTS import AutoTS
```

```python
model = AutoTS()
```

```python
model.fit(df,series_column_name="Price")
```

```python
df_res=pd.DataFrame(model.predict(start=pd.to_datetime('2020-08-01'), end=pd.to_datetime('2022-01-01')))
```

```python
df_res.rename(columns={0:"price_prediction"}, inplace = True)
```

```python
df_res
```

**** Dumping model for reuseabity ****

```python
import pickle 
pickle_out = open("model_used.pkl", mode = "wb") 
pickle.dump(model, pickle_out) 
pickle_out.close()
```

```python
from pickle import load
```
